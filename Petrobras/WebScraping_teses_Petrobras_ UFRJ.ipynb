```python
# import libraries
import urllib.request
from bs4 import BeautifulSoup
from selenium import webdriver
import selenium
import time
import pandas as pd
import numpy as np
import json
```

```python
empregados = pd.read_csv('Autores Petrobras_v2.csv')
```

```python
# specify the url
#urlpage = 'https://groceries.asda.com/search/yogurt' 
#url = 'https://pantheon.ufrj.br/handle/11422/1/browse?type=author&order=ASC&rpp=20&value=Bastos%2C+Leopoldo+Eurico+Gon√ßalves'
#print(urlpage)
# run firefox webdriver from executable path of your choice
#driver = webdriver.Firefox()
```

```python
links = []

for autor in empregados.iterrows():
    
    driver = webdriver.Firefox()
    #Separar nome e sobrenome do autor
    sobrenome = autor[1][0]
    nome = autor[1][1][1:]
    
    #Preparar nome e sobrenome para a url
    nome = '+'.join(nome.split())
    sobrenome = '+'.join(sobrenome.split())
    
    #preparar a url
    url = 'https://pantheon.ufrj.br/browse?type=author&value=' + sobrenome + '%2C+' + nome + '&filtername=author&filtertype=equals'
    print (url)
    #Fazer requisi√ß√£o e parsear o arquivo html

    driver.get(url)
    
    # find elements by xpath
    results = driver.find_elements_by_xpath("//*[@headers='t2']")
    
    # loop over results
    for result in results:
        link = result.find_element_by_tag_name('a')
        product_link = link.get_attribute("href")
        print(product_link)
        
        links.append(product_link)
    driver.quit()
```

```python
links = pd.read_csv("links_ufrj.csv", header=None)[0]
```

```python
teses_ufrj = {}

for url in links:

    driver = webdriver.Firefox()
    driver.get(url)

    label = driver.find_elements_by_xpath('//*[@class="metadataFieldLabel"]')
    value = driver.find_elements_by_xpath('//*[@class="metadataFieldValue"]')

    n_label = len(label)

    tese = {}
    for n in range(n_label):
        if (value[n].find_elements_by_tag_name('a') == [] or
            len(value[n].find_elements_by_tag_name('a')) == 1):
            tese[label[n].text] = value[n].text
        else:
            values = []
            for a in value[n].find_elements_by_tag_name('a'):
                values.append(a.text)
            tese[label[n].text] = values
        
    teses_ufrj[url] = tese
    driver.quit()
```

```python
with open('teses_petrobras_UFRJ.json', 'w') as fp:
    json.dump(teses_ufrj, fp)
```
